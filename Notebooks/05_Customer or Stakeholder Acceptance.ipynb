# 5. Customer / Stakeholder Acceptance

## 5.1 Conclusion

The rapid growth of the used car market in Saudi Arabia, driven by increasing demand for private vehicles and accelerated digitalization through platforms such as Syarah.com, creates a key challenge in determining fair vehicle prices due to the high variability in car conditions. To address this challenge, this project aims to develop a **regression-based machine learning model** capable of predicting the fair price of used cars based on important features such as brand, type, production year, engine size, mileage, region, and vehicle options. The model is intended to support objective and measurable decision-making for sellers, buyers, and platforms. Model performance is evaluated using the **Mean Absolute Error (MAE)** metric, with an initial target of MAE not exceeding 20% of the median price as a business-relevant accuracy indicator.

The **Saudi Arabia Used Cars** dataset from Kaggle, containing more than 8,000 used car records, was analyzed and prepared through data structure understanding, descriptive statistical analysis, *high-cardinality* identification, *missing value* inspection (particularly in *Engine_Size*), and numerical distribution analysis showing *right-skewed* patterns and significant outliers. The data cleaning process included conservative imputation, removal of irrelevant columns and duplicates, and outlier handling using the **IQR** method, followed by data type conversion and *feature engineering* such as **Car_Age** and **Mileage_per_Year** to enhance information representation.

The modeling phase was conducted end-to-end using an **80:20 train–test split** and an integrated **preprocessing pipeline** to prevent data leakage, separating treatments for numerical and categorical features. Several regression algorithms were evaluated using MAE-based **cross-validation**, including Linear Regression, Random Forest, Gradient Boosting, and XGBoost. The **XGBoost** model demonstrated the best and most stable performance, achieving a **test MAE of approximately 11,200 SAR**, which is **below 20% of the median vehicle price**, thus meeting the defined business accuracy target. The model was further improved through **hyperparameter tuning (GridSearchCV)**, resulting in a significant MAE reduction compared to the baseline.

Model interpretability was analyzed using **SHAP**, identifying *Engine_Size, Year, Mileage,* and *Car_Age* as the most dominant factors influencing price, followed by contributions from specific brands and vehicle types. The final model was saved as a trained artifact and implemented within a prediction pipeline ready for operational use, enabling consistent, measurable, and business-oriented fair market price estimation.

---

## 5.2 Limitations

* The model predicts **listed advertisement prices**, not final negotiated prices, meaning post-negotiation differences are not fully captured.
* Information regarding **technical vehicle condition** (e.g., engine condition, service history, or wear level) is limited, so the true car quality is not fully represented in predictions.
* Some data are dominated by **specific brands, years, and regions**, which may reduce prediction accuracy for vehicles with rare characteristics.
* The model learns **patterns from historical data** without accounting for time-based price changes or causal factors.
* The model does not yet include **performance monitoring or automatic retraining**, so prediction accuracy may decline if market conditions change.

---

## 5.3 Recommendations

* Integrate **actual transaction price data and negotiation history** to better reflect real market prices.
* Incorporate **more detailed vehicle condition information**, such as inspection results, service history, or accident records, to improve model accuracy.
* Develop **segment-specific models** (e.g., by brand, region, or price range) to maintain performance despite data imbalance.
* Build models that **incorporate temporal factors**, such as market trends, inflation, and seasonality, while beginning to explore simple causal relationships.
* Implement **regular model and data performance monitoring**, including periodic retraining, to maintain accuracy in real-world deployment.
