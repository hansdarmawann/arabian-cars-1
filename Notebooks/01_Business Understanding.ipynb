# 1. Business Understanding

## 1.1 Background

The used car market in Saudi Arabia has grown rapidly in recent years. This growth is driven by the increasing need for private vehicles for daily activities. In major cities such as Riyadh and Jeddah, many people prefer owning a car due to limited public transportation options. Additionally, comfort, time flexibility, and lifestyle factors have further increased interest in personal vehicles. As a result, the used car market has become one of the most active and competitive sectors in the automotive industry.

The advancement of digital technology has also accelerated the growth of the used car market. The emergence of online car marketplaces such as **Syarah.com** enables transactions to be conducted more easily and efficiently. Through these platforms, users can browse car options, compare prices, check vehicle conditions, and complete purchases online without visiting dealers in person. This digital process saves time and costs while improving transparency for both sellers and buyers.

**Syarah.com** is one of the largest online car marketplaces in Saudi Arabia, offering both new and used vehicles. Cars listed on the platform undergo comprehensive inspections and come with warranties, providing greater buyer confidence. Additionally, Syarah offers financing services and home delivery options. With a fully digital system, the car purchasing process becomes more practical and convenient for customers.

Despite these advantages, determining the right price for a used car remains a challenge. Each vehicle has unique characteristics such as production year, brand, model, and mileage, all of which significantly influence its selling price. Without a clear reference, sellers risk setting prices too high, making cars difficult to sell, or too low, leading to financial losses. On the other hand, buyers may overpay if they lack accurate market price information.

---

## 1.2 Business Objectives

This study aims to build a **regression-based machine learning model** capable of **estimating fair used car prices in Saudi Arabia**. Predictions are generated by considering key vehicle attributes such as **brand, model, production year, engine size, mileage, selling location, and available features**.

The developed model is expected to **support decision-making** for various stakeholders in the used car market:

* **For sellers**, the model can help determine **reasonable and competitive listing prices**, allowing cars to sell faster without underpricing.
* **For buyers**, the model can serve as a **reference to assess whether a car price is fair or overpriced** given its condition.
* **For marketplaces**, the model can provide a **more objective view of market prices**, supporting recommendation systems and price analytics features.

---

## 1.3 Key Business Questions

1. **What factors most strongly influence used car prices?**
   For example, whether production year, brand, mileage, or specific features have the greatest impact.

2. **How close are the model predictions to actual vehicle prices?**
   This question evaluates whether the model can estimate prices accurately and consistently.

3. **Are the model predictions reliable enough to be used as pricing references?**
   Both for sellers and buyers, as well as for pricing recommendation systems on used car platforms.

---

## 1.4 Success Criteria

The model is considered successful if it can **estimate used car prices closely and consistently relative to actual prices**. To evaluate this, a primary metric is used: **Mean Absolute Error (MAE)**.

**MAE** represents the **average price difference** between model predictions and actual prices, measured in SAR. In other words, MAE indicates **how far, on average, the model’s predictions deviate from real vehicle prices**. This metric is chosen because it is **easy to interpret**, less sensitive to extremely high-priced vehicles, and well aligned with the goal of fair price estimation.

As an initial benchmark, the model is considered sufficiently accurate if the **average prediction error does not exceed 20% of the median vehicle price**, with this threshold adjustable depending on business needs.
