# 1. Business Understanding

## 1.1 Latar Belakang

Pasar mobil bekas di Arab Saudi terus berkembang pesat dalam beberapa tahun terakhir. Hal ini didorong oleh meningkatnya kebutuhan masyarakat akan kendaraan pribadi untuk aktivitas sehari-hari. Di kota-kota besar seperti Riyadh dan Jeddah, banyak orang memilih memiliki mobil sendiri karena keterbatasan transportasi umum. Selain itu, faktor kenyamanan, fleksibilitas waktu, dan gaya hidup juga ikut meningkatkan minat terhadap mobil pribadi. Kondisi ini membuat pasar mobil bekas menjadi salah satu sektor otomotif yang sangat aktif dan kompetitif.

Perkembangan teknologi digital turut mempercepat pertumbuhan pasar mobil bekas. Kehadiran platform jual beli mobil secara online seperti **Syarah.com** memungkinkan proses transaksi dilakukan dengan lebih mudah dan efisien. Melalui platform ini, pengguna dapat melihat berbagai pilihan mobil, membandingkan harga, mengecek kondisi kendaraan, hingga menyelesaikan pembelian secara online tanpa harus datang langsung ke dealer. Proses digital ini membantu menghemat waktu dan biaya, sekaligus meningkatkan transparansi bagi penjual dan pembeli.

**Syarah.com** merupakan salah satu platform jual beli mobil online terbesar di Arab Saudi yang menyediakan mobil baru maupun bekas. Mobil yang dijual telah melalui proses inspeksi menyeluruh dan dilengkapi dengan garansi, sehingga pembeli merasa lebih aman. Selain itu, Syarah juga menawarkan layanan pembiayaan dan pengantaran mobil langsung ke rumah. Dengan sistem yang sepenuhnya digital, proses pembelian mobil menjadi lebih praktis dan nyaman bagi masyarakat.

Meskipun demikian, menentukan harga mobil bekas yang tepat masih menjadi tantangan. Setiap mobil memiliki kondisi yang berbeda-beda, seperti tahun produksi, merek, tipe, dan jarak tempuh, yang sangat memengaruhi harga jualnya. Tanpa acuan yang jelas, penjual berisiko memasang harga terlalu tinggi sehingga mobil sulit terjual, atau terlalu rendah sehingga merugi. Di sisi lain, pembeli juga berpotensi membayar lebih mahal dari nilai pasar jika tidak memiliki informasi harga yang akurat.

## 1.2 Tujuan Bisnis

Penelitian ini bertujuan untuk membangun **model machine learning berbasis regresi** yang dapat **memperkirakan harga mobil bekas yang wajar di Arab Saudi**. Prediksi dilakukan dengan mempertimbangkan berbagai informasi penting dari kendaraan, seperti **merek, tipe mobil, tahun produksi, ukuran mesin, jarak tempuh, lokasi penjualan, serta fitur yang dimiliki**.

Model yang dibangun diharapkan dapat **membantu pengambilan keputusan** bagi berbagai pihak yang terlibat di pasar mobil bekas, antara lain:

* **Bagi penjual**, model dapat membantu menentukan **harga jual yang masuk akal dan kompetitif**, sehingga mobil lebih cepat terjual tanpa harus menjual terlalu murah.
* **Bagi pembeli**, model dapat digunakan sebagai **acuan untuk menilai apakah harga mobil sudah sesuai atau terlalu mahal** dibandingkan dengan kondisi mobil tersebut.
* **Bagi platform jual beli**, model ini dapat dimanfaatkan untuk memberikan **gambaran harga pasar yang lebih objektif**, serta mendukung fitur rekomendasi dan analisis harga secara lebih terukur.

## 1.3 Pertanyaan Bisnis Utama

1. **Faktor apa saja yang paling memengaruhi harga mobil bekas?**
   Misalnya apakah tahun produksi, merek, jarak tempuh, atau fitur tertentu memiliki pengaruh paling besar terhadap harga.

2. **Seberapa dekat hasil prediksi model dengan harga mobil sebenarnya?**
   Pertanyaan ini bertujuan untuk melihat apakah model mampu memperkirakan harga dengan cukup akurat dan konsisten.

3. **Apakah hasil prediksi model cukup layak digunakan sebagai acuan harga?**
   Baik sebagai referensi bagi penjual dan pembeli, maupun sebagai dasar rekomendasi harga di platform jual beli mobil bekas.

## 1.4 Definisi Keberhasilan

Model dikatakan berhasil jika mampu **memperkirakan harga mobil bekas dengan cukup dekat dari harga aslinya dan konsisten**. Untuk menilai hal tersebut, digunakan satu ukuran utama yaitu **Mean Absolute Error (MAE)**.

**MAE** menunjukkan **rata-rata selisih harga** antara hasil prediksi model dan harga sebenarnya dalam satuan Rupiah/SAR. Dengan kata lain, MAE menggambarkan **seberapa jauh rata-rata prediksi model meleset dari harga asli mobil**. Metrik ini dipilih karena **mudah dipahami**, tidak terlalu terpengaruh oleh harga mobil yang sangat mahal, dan sesuai dengan kebutuhan penentuan harga yang wajar.

Sebagai patokan awal, model dianggap sudah cukup baik apabila **rata-rata selisih prediksi harga tidak melebihi 20% dari harga tengah (median) mobil**, dengan ketentuan ini dapat disesuaikan kembali sesuai kebutuhan bisnis.
