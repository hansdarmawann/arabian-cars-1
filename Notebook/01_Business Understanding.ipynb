# 1. Business Understanding

## 1.1 Latar Belakang

Pasar mobil bekas di Arab Saudi sedang mengalami pertumbuhan yang sangat pesat dalam beberapa tahun terakhir. Kenaikan ini dipicu oleh meningkatnya kebutuhan masyarakat terhadap kendaraan pribadi untuk menunjang mobilitas sehari-hari. Banyak penduduk di kota besar seperti Riyadh dan Jeddah yang lebih memilih memiliki mobil sendiri karena transportasi umum masih terbatas. Faktor lain seperti kenyamanan, fleksibilitas waktu, dan status sosial juga berperan dalam meningkatkan permintaan. Akibatnya, sektor mobil bekas menjadi salah satu segmen otomotif yang paling dinamis di negara tersebut.

Selain faktor permintaan, perkembangan teknologi digital juga berperan besar dalam mempercepat pertumbuhan pasar mobil bekas di Arab Saudi. Platform jual beli online seperti **Syarah.com** memudahkan penjual dan pembeli untuk bertransaksi tanpa harus datang ke dealer secara langsung. Melalui sistem daring, pengguna dapat membandingkan harga, memeriksa kondisi mobil, hingga menyelesaikan proses pembelian secara penuh secara online. Hal ini memberikan efisiensi waktu dan biaya yang signifikan bagi kedua belah pihak, sekaligus meningkatkan transparansi dalam transaksi. Tidak heran jika platform semacam ini kini menjadi salah satu pendorong utama perkembangan ekosistem otomotif digital di negara tersebut.

**Syarah.com** sendiri merupakan platform jual beli mobil online terkemuka asal Arab Saudi yang melayani transaksi mobil baru maupun bekas. Setiap mobil yang dijual telah melewati inspeksi ketat lebih dari 200 titik dan dilengkapi dengan garansi hingga satu tahun untuk menjamin kualitasnya. Selain itu, Syarah menyediakan layanan pembiayaan yang fleksibel serta opsi pengantaran mobil langsung ke rumah pembeli. Dengan proses yang sepenuhnya digital, pengguna dapat membeli mobil dengan mudah, aman, dan nyaman melalui situs atau aplikasi resmi mereka. Berbasis di Riyadh, Syarah kini dikenal sebagai salah satu marketplace otomotif terbesar dan paling terpercaya di Arab Saudi.

Namun, besarnya variasi kondisi kendaraan membuat **penentuan harga yang wajar** menjadi tantangan tersendiri. Setiap mobil memiliki karakteristik unik berdasarkan tahun produksi, merek, tipe, serta jarak tempuh yang memengaruhi nilai jualnya. Tanpa alat bantu yang tepat, penjual bisa saja menetapkan harga terlalu tinggi sehingga mobilnya sulit terjual. Sebaliknya, harga yang terlalu rendah dapat menyebabkan kerugian finansial. Bagi pembeli, ketidaktahuan terhadap nilai pasar juga dapat membuat mereka membayar lebih mahal dari seharusnya.

## 1.2 Tujuan Bisnis

Melalui pendekatan **causal-based forecasting (regresi)**, penelitian ini bertujuan untuk membangun **model machine learning** yang mampu **memprediksi harga wajar mobil bekas di Arab Saudi** berdasarkan berbagai fitur penting, seperti merk, tipe, tahun produksi, ukuran mesin, jarak tempuh, region, serta opsi fitur yang dimiliki kendaraan. Model ini diharapkan dapat **mendukung pengambilan keputusan** bagi berbagai pemangku kepentingan di pasar mobil bekas, yaitu: 
- Bagi **penjual**, model dapat digunakan untuk menentukan **harga listing yang kompetitif** sehingga peluang terjual lebih cepat meningkat tanpa mengorbankan potensi keuntungan. 
- Bagi **pembeli**, model membantu menilai apakah harga sebuah mobil **tergolong wajar atau cenderung overpriced** dibandingkan kondisi dan karakteristiknya. 
- Sementara bagi **platform**, model ini dapat dimanfaatkan untuk memberikan **rekomendasi dan insight harga pasar** secara lebih objektif dan terukur.

## 1.4 Definisi Keberhasilan

Keberhasilan model diukur berdasarkan kemampuan prediksinya dalam memperkirakan harga mobil bekas secara akurat dan konsisten. Evaluasi dilakukan menggunakan **metric utama** berupa **MAE (Mean Absolute Error)** dan **MAPE (Mean Absolute Percentage Error)**.

- **MAE** mengukur rata-rata selisih absolut antara harga aktual dan harga hasil prediksi, sehingga memberikan gambaran seberapa besar deviasi model terhadap nilai sebenarnya dalam satuan mata uang.
- **MAPE** mengukur persentase kesalahan relatif terhadap harga aktual, yang menunjukkan seberapa besar error model dalam konteks proporsional.

Sebagai target awal, model dianggap berhasil apabila mencapai **MAE <= 20% dari median harga**, dengan ambang batas ini dapat disesuaikan sesuai kebutuhan bisnis dan karakteristik data.
