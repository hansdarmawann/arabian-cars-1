# 1. Business Understanding (TDSP Step 1)

## 1.1 Latar Belakang

Pasar mobil bekas di Arab Saudi berkembang pesat, didorong oleh:

- Tingginya permintaan kendaraan untuk mobilitas harian.
- Maraknya platform jual beli online seperti **Syarah.com** yang mempertemukan penjual dan pembeli.
- Variasi kondisi kendaraan (tahun, merk, tipe, jarak tempuh, dsb.) yang membuat **penentuan harga wajar** menjadi tantangan.

Tanpa alat bantu, penjual bisa memasang harga terlalu tinggi (sulit laku) atau terlalu rendah (merugikan). Pembeli pun berisiko membayar lebih mahal dari harga pasar.

---

## 1.2 Tujuan Bisnis

Membangun **model machine learning** yang mampu:

1. **Memprediksi harga wajar** mobil bekas di Arab Saudi berdasarkan fitur seperti merk, tipe, tahun, ukuran mesin, jarak tempuh, region, dan opsi fitur.
2. **Mendukung pengambilan keputusan**:
   - Penjual: menentukan harga listing yang kompetitif.
   - Pembeli: menilai apakah harga sebuah mobil wajar atau overpriced.
   - Platform: memberikan rekomendasi / insight harga pasar.

---

## 1.3 Pertanyaan Bisnis Utama

1. Faktor apa saja yang paling berpengaruh terhadap harga mobil bekas?
2. Seberapa akurat model dalam memprediksi harga?
3. Apakah model cukup baik untuk dijadikan referensi harga di platform?

---

## 1.4 Definisi Keberhasilan

- **Metric utama:**
  - **MAE (Mean Absolute Error)** — rata-rata selisih absolut antara harga aktual dan prediksi.
  - **MAPE (Mean Absolute Percentage Error)** — persentase error relatif terhadap harga aktual.
- Target (dapat disesuaikan):
  - MAE ≤ 10% dari median harga.
  - MAPE ≤ 20%.
